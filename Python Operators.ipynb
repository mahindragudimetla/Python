4.  <	Less than	x < y

# **4. Logical Operators:-**
Logical operators are used to combine conditional statements:

Op-----------------Description-----------------------------------Example
and------Returns True if both statements are true--------x < 5 and  x < 10
or-------Returns True if one of the statements is true-----x < 5 or x < 4
not----Reverse the result, returns False if the result is true----not(x < 5 and x < 10)
```
print(4<5 and 3<1)  
print(4<5 or 3<1)  
print(not(4<5 and 3<1))
```

# **5.Identity Operators:-**
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:

1.  is :-Returns True if both variables are the same object  i.e. a is b returns true if a and b are same
2.  is not :- Returns True if both variables are not the same object  i.e. a is not b returns true if a and b are not same

# **6.Membership Operators:-**
Membership operators are used to test if a sequence is presented in an object:

1.  in :-Returns True if a sequence with the specified value is present in the object	 i.e x in y returns true if the value of x is present in y
2.  not in :-Returns True if a sequence with the specified value is not present in the object	 i.e x in y returns true if the value of x is not present in y

1. What is Byte?
Bytes are the smallest addressable unit of memory in many computer architectures. They are used in many areas, including:
Programming: Used as a unit of measurement for memory and storage, as well as a means of representing data
Cryptography: Used to represent encrypted data, keys, and other important information
Computer storage is typically measured in byte multiples, such as gigabytes (GB) and terabytes (TB).

Note:- A byte is equal to 8 bits

2. What is a bit?
A bit (binary digit) is the smallest unit of data that a computer can process and store. A bit is always in one of two physical states, similar to an on/off light switch. The state is represented by a single binary value, usually a 0 or 1. However, the state might also be represented by yes/no, on/off or true/false.

3. What is Kbs?
A kilobyte (KB) is a unit of digital information storage that is equal to about 1,000 bytes. It is used to measure the size of files, storage capacity, and data transfer rates.

Note:- one kilobyte is 1000 bytes  and 1 megabyte is equivalent to 1,000 kilobytes.

3. What is Mb's:-
One megabyte, or MB, is equivalent to 1024 kilobytes, or KB, in digital units of measurement. About a million bytes make up a megabyte (or about 1000 kilobytes). The megabyte is a multiple of the digital informational unit byte. The suggested unit symbol for it is MB

4.  What is Gb's:-
GB stands for gigabyte, which is a unit of measurement for digital information storage capacity. It's commonly used in computing, technology, programming, and communications.

Note:- One gigabyte is equal to: 1 billion bytes of data, 1,024 megabytes (MB), and 0.001 terabytes (TB).

5. What is a Tb's:-
A terabyte (TB) is a unit of digital data storage that is equal to one trillion bytes. It is a relatively large unit of data volume and is commonly used to measure the capacity of storage devices and hard drives.

Note:- Terabyte is 1,000,000,000,000 bytes, or 1,000 gigabytes

Divide the decimal number by 2

Write down the remainder, which is either 0 or 1

Repeat the process using the quotient from the previous division as the new

Read the remainders in reverse order to get the binary representation
