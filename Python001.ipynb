# **IV. Data Types:-**
There are two types in data tupes
1. Basic
2. Advances ( will be discussed at a later time)

**1.Basic Data types:-**
It consists of 4 data types:-

1. int
2. float
3. complex
4. Boolean

**int:-**
int, or integer, is a whole number, positive or negative, without decimals, of unlimited length.

E.g:-1,1345,-41313, etc.....

**float:-**
float, or "floating point number" is a number, positive or negative, containing one or more decimals.


E.g:- 1.213, 0.313, -31.123,,, etc......

**complex:-**
Complex numbers are a combinations of both intergers/float and imaginary value.

Imaginary value are written with a "j" as the imaginary part:

E.g:- 1+0j, 2+345j, 31.431+31j, etc......

**Boolean:-**
Python Boolean type is one of the built-in data types provided by Python, which represents one of the two values i.e. True or False. Generally, it is used to represent the truth values of the expressions.

Boolean value can be of two types only i.e. either True or False. The output <class ‘bool’> indicates the variable is a Boolean data type.

# **V. Type Conversions(casting):-**
Type conversions are used to convert one data type into another data type using int, float, complex functions.

These are following way we can change them
1. int to float:-

```
a=7
print(float(a)) #out=7.0
```
2. float to int:-
```
a=7.0
print(int(a)) #out=7
```
3. int to complex:-
```
a=7
print(complex(a)) #out=7+0j
```
4. float to complex:-
```
a=7.5
print(complex(a)) #out=7.5+0j
```
**Note:-**
However we cannot convert a complex into a integer or float because of the imaginary number.

**Index:-**
An index is used to print a certain characted/element from a group of elements.

Like many other popular programming languages, strings in Python are arrays of bytes representing unicode characters.

However, Python does not have a character data type, a single character is simply a string with a length of 1.

Square brackets can be used to access elements of the string

Index is divided into 2 types:-

1.  Positive Indexing:-
In Python, positive zero-based indexing is the fundamental method for accessing iterable items.
As a result, an index starting at 0 may refer to any element in the iterable.
The first element in zero-based indexing has an index of 0, the second has an index of 1, and so on.
It goes from right to left
```
a="Example"
#  0123456 are the positive index values
print(a[4]) #out:-p
```
2.  Negative Indexing:-
If we wish to print the components starting at the end, we may also use negative indexes. Additionally, by specifying the index number with a negative sign, we may carry out negative tuple indexing (-). As a result, this suggests that the compiler starts thinking about the elements in reverse order, beginning with the element that comes last in the tuple.
```
a="Example"
print(a[-6]) #out:- x
print(a[-3]) #out:- p
```

**Slicing:-**
Slicing is used to print a certain group of elements using the index values.

Syntax:- [starting index no.:ending index no.+1]
```
a="Examples"
print(a[3:7])
#out:-mple
#prints from 3 to the 6th elements
```
```
a="Examples"
print(a[-7:-2])
#out:-xampl
#prints from -7 to the -3 elements
```

**String Format:-**

1.  .format() method:-
The format() method formats the specified value(s) and insert them inside the string's placeholder.

The placeholder is defined using curly brackets: {}. Read more about the placeholders in the Placeholder section below.

The format() method returns the formatted string
```
name="mahi"
age="21"
print("My name is {} and i am {} years old".format(name,age))
#out:-My name is mahi and i am 21 years old
```

2.  f-string method:-
F-string allows you to format selected parts of a string.
```
name="mahi"
age=21
print(f"My name is {name},I am {age} years old")
```
