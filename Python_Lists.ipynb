# **VIII.Lists:-**

Lists are used to store multiple items in a single variable.(different datatypes)

Lists are one of 4 built-in data types in Python used to store collections of data, the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.

List items are ordered, changeable, and allow duplicate values.

List items are indexed, the first item has index [0], the second item has index [1] etc.

When we say that lists are ordered, it means that the items have a defined order, and that order will not change.

If you add new items to a list, the new items will be placed at the end of the list.

Lists are created using square brackets:
```
thislist = ["apple", "banana", "cherry"]
print(thislist)
#out:-["apple", "banana", "cherry"]
```

Here we can use Indexing to extract a single element from a list.

E.g:-
```
print(thislist[2])
#out:- "cherry"
```

We can also use slicing to extract multiple elements
```
print(thislist[1:2])
#out:-["apple","banana"]
```


** List Methods:-**
1. append():- append() is used to add mew elements at the end of the list.
```
a=[1,2,3,4,5]
a.append(7)
print(a)
#out:-[1,2,3,4,5,7]
```

2. extend():-Used to add multiple elements.
```
fruits = ['apple', 'banana', 'cherry']
cars = ['Ford', 'BMW', 'Volvo']
fruits.extend(cars)
print(fruits)
#out:-['apple', 'banana', 'cherry','Ford', 'BMW', 'Volvo']
```

3. len():-Used to find the length of the list.
```
a=[1,2,3,4,5,6,7,8,9]
print(len(a))
#out:-9
```
4. count():- Used to count the number of time a element is repeated in the list.
```
a=[1,2,3,3,2,1,31,234,2,3]
a.count(3)
#out:-3  since 3 is repeated 3 times in the list
```
5. index():-Used to find the index value of a element.
```
a=[1,2,3,4,5,6,7]
a.index(4)
#out:-3 since 3 is the index value of 4
```
6. pop():- Used to remove a element from a list. It usually deletes the last element from a list by default but if we mention the element that needs to be removed it will delete that
```
a=[1,2,3,4,5]
a.pop()
print(a)
#out:-[1,2,3,4]  it deletes 5 since we didn't mention anything
# or
a.pop(3)
print(a)
#out:-[1,2,4] since we mentioned 3 to pop it deletes the 3 and prints the rest
```
