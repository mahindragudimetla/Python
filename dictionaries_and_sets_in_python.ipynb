# **X. Dictionaries:-**
Dictionaries are used to store data values in key:value pairs.

A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

Dictionaries are written with curly brackets, and have keys and values pairs.

The value can only changed using the keys as index

If there is a duplicate value with same key then the later value will only be considered as it will think that we are updating the key.
```
dic= {1:'a', 2:'b', 3:'c'}
#Here lets do indexing
#In dictionaries indexing can only be done using the keys
print(dic[1])
#Out:-'a'
print(dic[3])
#out:-'c'
```
**Updating in python:-**
In  dictionary we can use the update to update an already existing value of a key to another new one
```
dic= {1:'a', 2:'b', 3:'c'}
dic[1]="hello"
print(dic)
#out:-{1:"hello",2:'b',3:'c'}
```

**Dictionary methods:-**

1.  keys():-
Used to print the keys of the dictionary.
E.g:-
```
a={1:'a', 2:'b', 3:'c'}
a.keys()
#o/p:- [1,2,3]
```

2.  values:-
Used to print the values of the dictionary.
E.g:-
```
a={1:'a', 2:'b', 3:'c'}
a.values()
#o/p:- ['a','b','c']
```

3.  items:-
Used to print both the keys and values of the dictionary.
E.g:-
```
a={1:'a', 2:'b', 3:'c'}
a.items()
# o/p:- dict_items([(1, 'a'), (2, 'b'), (3, 'c')])
```

4.  get():-
To print the value of the specified key
E.g:-
```
a={1:'a', 2:'b', 3:'c'}
a.get(1)
#o/p:- 'a'
```

5.  update():-
Updates the dictionary with the specified key-value pairs
E.g:-
```
a={1:'a', 2:'b', 3:'c'}
a.update({1:"hello"})
a.update({4:'d})
print(a)
#o/p:- {1:'hello', 2:'b', 3:'c', 4:'d'}
```

6.  pop():-
Removes the element with the specified key
```
a={1:'a', 2:'b', 3:'c'}
a.pop(1)
print(a)
#o/p:- {2:'b', 3:'c'}
```

7.  popitem():-
Removes the last inserted key-value pair by default
```
a={1:'a', 2:'b', 3:'c'}
a.popitem()
print(a)
#o/p:- {1:'a', 2:'b'}
```



# **XI. SETS:-**
Sets are used to store multiple items in a single variable.

Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Tuple, and Dictionary, all with different qualities and usage.

A set is a collection which is unordered, unchangeable*, and unindexed.

Sets are written with curly brackets.

Set items are unordered, unchangeable, and do not allow duplicate values.

**Set Methods:-**

1.  union():-
To get a combined set of the 2 and
Represented using """   |   """
E.g:-
```
a={1,2,3,4}
b={3,4,5,6,7}
print(a|b)
#o/p:- {1,2,3,4,5,6,7}
```

2. intersection:-To print the common elements between the 2 sets and
Represented using  &
E.g:-
```
a={1,2,3,4}
b={3,4,5,6,7}
print(a&b)
#o/p:- {3,4}
```

3.  discard():- Remove the specified item
E.g:-
```
a={1,2,3,4,5,6}
a.discard(3)
#o/p:- {1,2,4,5,6}
```

4.  difference of 2 sets:-
Returns a set containing the difference between two or more sets
```
a={1,2,3,4,5}
b={3,4,5,6,7}
print(a-b)
#o/p:-{1,2}
print(b-a)
#o/p:-{6,7}
```

5.  symmetric difference:-
Returns a set with the symmetric differences of two sets and
represented using     ^  
```
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
print(a^b)
#o/p:-{1,2,3,7,8,9}
```
6.  add():-
Adds an element to the set
```
a={1,2,3,5,4,6}
a.add(21)
a
#o/p:- {1,2,3,4,5,6,21}
```

7.  update():-Update the set with the union of this set and others
```
a={1,2,3,4,5}
a.update({3,4,5,6,7,8,9,0})
a
#o/p:-{1,2,3,4,5,6,7,8,9,0}
```
